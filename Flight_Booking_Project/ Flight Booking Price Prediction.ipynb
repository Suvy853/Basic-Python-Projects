```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd

# Create a list of numeric columns excluding 'price'
col_list = []
for col in df.columns:
    if (df[col].dtype != "object") and (col != "price"):  # Corrected closing parenthesis
        col_list.append(col)

# Create the DataFrame for VIF calculation
x = df[col_list]
vif_data = pd.DataFrame()
vif_data["feature"] = x.columns
vif_data["VIF"] = [variance_inflation_factor(x.values, i)
                   for i in range(len(x.columns))]

print(vif_data)
```

```python
# Splitting the dataset into features (x) and target (y)
y = df['price']  # Target variable
x = df.drop(columns=['price'])  # Features

# Splitting into training and testing datasets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Verifying shapes of the splits
print("x_train shape:", x_train.shape)
print("x_test shape:", x_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
# Assuming 'price' is the target column
y = df['price']
x = df.drop(columns=['price'])

# Handle non-numeric columns
from sklearn.preprocessing import LabelEncoder
for col in x.select_dtypes(include=['object']).columns:
    encoder = LabelEncoder()
    x[col] = encoder.fit_transform(x[col])

# Verify all columns are numeric
print(x.dtypes)

# Train-test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Standard scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)

# Fit a Linear Regression model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)

# Evaluate the model
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
rmse = mean_squared_error(y_test, y_pred, squared=False)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")
print(f"R Score: {r2:.2f}")
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

# Calculate metrics
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100

# Print metrics
print(f"R Score: {r2:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")
```

```python
# Inspect data range
print(f"y_test range: {y_test.min()} to {y_test.max()}")
print(f"y_pred range: {y_pred.min()} to {y_pred.max()}")

# Define thresholds
min_threshold = 0  # Example: No negative prices
max_threshold = 50000  # Example: Set a realistic upper bound

# Clip values
y_test = y_test.clip(lower=min_threshold, upper=max_threshold)
y_pred = pd.Series(y_pred).clip(lower=min_threshold, upper=max_threshold)

# Plot after clipping
import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(y_test, label="Actual", bins=30, kde=True, color='blue')
sns.histplot(y_pred, label="Predicted", bins=30, kde=True, color='red')
plt.title("Distribution of Actual vs Predicted Prices")
plt.xlabel("Price")
plt.ylabel("Density")
plt.legend(["Actual", "Predicted"])
plt.grid(True)
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
import numpy as np

# Initialize and fit the Decision Tree Regressor
dt = DecisionTreeRegressor()
dt.fit(x_train, y_train)

# Make predictions
y_pred = dt.predict(x_test)

# Calculate R score
r2 = r2_score(y_test, y_pred)
print(f"R Score: {r2:.2f}")

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error (MAE): {mae:.2f}")

# Calculate Mean Absolute Percentage Error (MAPE)
mape = mean_absolute_percentage_error(y_test, y_pred)
print(f"Mean Absolute Percentage Error (MAPE): {mape * 100:.2f}%")  # Convert to percentage

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error (MSE): {mse:.2f}")

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
import numpy as np

# Initialize and fit the Random Forest Regressor
rf = RandomForestRegressor(random_state=42)
rf.fit(x_train, y_train)

# Make predictions
y_pred = rf.predict(x_test)

# Calculate R score
r2 = r2_score(y_test, y_pred)
print(f"R Score: {r2:.2f}")

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error (MAE): {mae:.2f}")

# Calculate Mean Absolute Percentage Error (MAPE)
mape = mean_absolute_percentage_error(y_test, y_pred)
print(f"Mean Absolute Percentage Error (MAPE): {mape * 100:.2f}%")  # Convert to percentage

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error (MSE): {mse:.2f}")

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
```
